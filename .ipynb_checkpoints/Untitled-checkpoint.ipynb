```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. zadatak

```python
scales = [1.0, 0.25, 4.0]
w = 2
x = np.linspace(-8, 10, 100)

for s in scales:
    y = 1 / (1 + abs(x - w) / s)
    plt.plot(x, y)
    plt.title(f'S: {s}')
    plt.show()
```
