```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. zadatak

```python
scales = [1.0, 0.25, 4.0]
w = 2
x = np.linspace(-8, 10, 100)

for s in scales:
    y = 1 / (1 + abs(x - w) / s)
    plt.plot(x, y)
    plt.title(f'S: {s}')
    plt.show()
```

```python
import pandas as pd
from mlutils import plot_2d_clf_problem
```

```python
def class_from(y1: int, y2: int, y3: int):
    if y1 == 1:
        return 1
    if y2 == 1:
        return 2
    if y3 == 1:
        return 3

df = pd.read_csv('dataset.txt', sep='\t', names=['X1', 'X2', 'Y1', 'Y2', 'Y3'])
df['Class'] = df['Y1'] + 2 * df['Y2'] + 3 * df['Y3']
df = df.drop(columns=['Y1', 'Y2', 'Y3'])
X = df.values[:,0:2]
y = df.values[:,2]
plot_2d_clf_problem(X, y)
plt.title('Correct values')
```
